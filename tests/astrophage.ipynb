```python
def count_sample_references(df):
    # Group by 'sample' and aggregate
    result = df.groupby('sample').agg(
        unique_references=('reference', 'nunique'),  # Count unique references
    ).reset_index()
    
    return result

result_df = count_sample_references(df)
print(result_df)
```
